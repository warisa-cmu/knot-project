```python
import matplotlib.pyplot as plt
import torchvision.transforms as transforms
from PIL import Image

def image_augmentation_and_display(image_path):
    image = Image.open(image_path)
    augmentation = transforms.Compose([
        transforms.RandomResizedCrop(size=(1000, 1000), scale=(2.0, 2.0)),
        transforms.RandomAffine(degrees=30, translate=(0.1, 0.1)),
        transforms.ColorJitter(brightness=0.5, contrast=0.5),
        transforms.ToTensor()
    ])
    augmented_image = augmentation(image)
    
    # Convert tensor back to PIL image for display
    augmented_image_pil = transforms.ToPILImage()(augmented_image)
    
    # Display the image
    plt.imshow(augmented_image_pil)
    plt.axis('off')
    plt.savefig("output.jpg", bbox_inches='tight', pad_inches=0)

# Example usage (uncomment to use):
image_augmentation_and_display('img/cc1-2.tiff')
```
